# Pandas

__pandas__ es una libreria de código abierto con herramientas para análisis de datos usando Python.

Pandas es, principalmente, una clase llamada _dataframe_. Igual que el arreglo de __numpy__ practicamente no usaremos funciones (excepto para crear un nuevo _dataframe_ y para algunas operaciones) sino los métodos que dicha clase tiene. Podemos crear un _dataframe_ vacía, pero lo normal es que importemos los datos de algún lugar.

__pandas__ nos ofrece herramientas para interactuar con datos guardados en diversos tipos de archivos.

|Format Type |	Data  Description |	Reader | Writer |
|---|---|---|---|
|text | CSV | read_csv | to_csv |
|text |JSON |read_json |to_json|
|text |HTML |read_html |to_html|
|text |Local clipboard |read_clipboard |to_clipboard|
|binary |MS Excel |read_excel |to_excel|
|binary |HDF5 Format |read_hdf |to_hdf|
|binary |Feather Format |read_feather |to_feather|
|binary |Msgpack |read_msgpack |to_msgpack|
|binary |Stata |read_stata |to_stata|
|binary |SAS |read_sas | | 
|binary |Python Pickle Format |read_pickle |to_pickle|
|SQL |SQL |read_sql |to_sql|
|SQL |Google Big Query |read_gbq |to_gbq|

Como ejemplo, abriremos un archivo con las calificaciones de un grupo.

```python
import pandas as pd
```

```python
df = pd.read_csv("ListaAlumnos.csv", index_col="Numero")
df
```

`<dataframe>.head()` y `<dataframe>.tail()` nos permite ver los primeros y últimos datos de nuestro _dataframe_.

```python
df.head()
```

```python
df.tail(3)
```

__pandas__ incluye sencillos para obtener indicadores estadisticos automáticamente. Uno de ellos es `<dataframe>.describe()`.

```python
df.describe()
```

__Nota__: `describe()` automáticamente quita los datos no numéricos.

Para calcular la matriz de coorelaciones podemos usar `<dataframe>.corr()`.

```python
df.corr()
```

Podemos ver solo un campo usando el punto y llamando el campo como una propiedad de nuestro _dataframe_:

```python
df.Examen1
```

O de la misma manera que haríamos en un diccionario:

```python
df['Examen1']
```

O extraer los datos a un arreglo de numpy usando la propiedad `<dataframe>.values`:

```python
df.Examen1.values
```

Para extraer varios campos usamos doble corchete:

```python
df[['Examen1', 'Examen2', 'Examen3']]
```

```python
df[['Examen1', 'Examen2', 'Examen3']].values
```

__pandas__ nos permite añadir (o modificar) campos ya sea con los operadores incluidos:

```python
df['PromTareas'] = df[['Tarea1', 'Tarea2', 'Tarea3']].mean(axis=1)
df
```

o con operadores externos:

```python
df['PromExamenes'] = df[['Examen1', 'Examen2', 'Examen3']].values.sum(axis=1) / 3
df
```

```python
df['Promedio'] = df.PromTareas * 0.3 + df.PromExamenes * 0.7
df[['Expediente', 'Nombre', 'Apellido', 'Promedio']]
```

Tambien nos permite hacer búsquedas:

```python
df[df["Apellido"]=="Hernandez"]
```

Para dos variables (usando parentesis __()__ en cada criterio y operadores __&__ (and) y __|__ (or).

```python
df[(df["Promedio"]>=80) & (df["Promedio"]<90.0)]
```

Podemos incluso modificar un campo __in place__.

```python
puntos_extra_proyecto = [0,1,0,0,0,0,0,1,0,1,1,0,1,0,0,1]

df.Promedio += puntos_extra_proyecto
df[['Expediente', 'Nombre', 'Apellido', 'Promedio']]
```

Podemos generar gráficas de nuestros datos con el método `plot()` (que usa internamente __matplotlib__).

```python
%matplotlib inline
df.Promedio.plot()
```

Ejemplo: desempeño de los alumnos en sus examenes a lo largo del semestre.

```python
for i, alumno in df.iterrows():
    ax = alumno[['Examen1', 'Examen2', 'Examen3']].plot(figsize=(10,7), legend=True)
ax.legend(df.Nombre)
```
