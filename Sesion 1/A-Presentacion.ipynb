### Programaci√≥n orientada a objetos:
Definimos una clase:

```python
class Ventana():
    """Clase Ventana. Tiene una √°rea definida, puede estar abierta o cerrada y puede o no 
    tener mosquitero"""
    def __init__(self, area, abierta, mosquitero = False):
        """Crea una ventana nueva. Necesita los valores del √°rea (metros cuadrados), 
        si est√° abierta o no y opcionalmente se puede delcarar si tiene o no mosquitero"""
        self.area = area
        self.abierta = abierta
        self.mosquitero = mosquitero
    
    def abrir(self):
        """Abre la ventana."""
        self.abierta = True

    def cerrar(self):
        """Cierra la ventana."""
        self.abierta = False
```

```python
help(Ventana)
```

Y ahora creamos una instancia (objeto) tipo Ventana:

```python
# Creamos la ventana_frontal:
ventana_frontal = Ventana(0.4*0.4, True)
# Utilizamos sus m√©todos y consultamos sus propiedades
print('La ventana frontal, ¬øest√° abierta?')
print(ventana_frontal.abierta)
print('¬øque √°rea tiene?')
print(ventana_frontal.area)
ventana_frontal.cerrar()
print('Ahora, ¬øest√° abierta?')
print(ventana_frontal.abierta)
```
